<h3>GPS coordinates

<h4>Calling API nominatim and start collecting cities GPS coordinates and store it in a list that will be used when calling the weather API (see next steps)

```python
#the following list is used to store GPS coordinates of cities
cities_locations=[]
#Nominatim function is used in this project to resolve Gps coorf=dinates of some cities not returned by the API like "Gorges du Verdon"
geolocator = Nominatim(user_agent="my_user_agent")
for i in cities :
    #print(i)
    city =i
    country ="France"
    #loc = geolocator.geocode(city+','+ country)
    #print("latitude is :-" ,loc.latitude,"\nlongtitude is:-" ,loc.longitude)
    
    url = "http://nominatim.openstreetmap.org/search?country=france&city={}&format=json".format(i.replace(" ","+"))
    print(url)
    r = requests.get(url)
    #r.text[100]
    df = r.json()
    
    #check if the API accept the get resquest
    if ( (r.status_code == 200)  ):
        #check if the API return a matching result
        if (len(df) >0):
            cities_locations.append([i,df[0]["lat"],df[0]["lon"] ])
        #if the API don't return a matching result we get a try with Nominatim function
        else:
            city =i
            country ="France"
            loc = geolocator.geocode(city+','+ country)
            cities_locations.append([i,str(loc.latitude),str(loc.longitude)])            
    else:
        cities_locations.append([i,"NA","NA" ])
print("fin 2")
```

<h3>Openweather forecast

<h4>Calling API openweather and start collecting cities weather forecast 

```python
import os
idx=""
cities_weather=[]
for i in cities_locations : 
    url =f"https://api.openweathermap.org/data/2.5/forecast?lat={i[1]}&lon={i[2]}&exclude=hourly,current,minutely&units=metrics&appid=522a267f52a0dfc5f1ec70f2a94be2af"
    #print("url====>", url)
    r = requests.get(url)
    res0 =[]
    res1 =[]
    #check if the API accept the get resquest
    if r.status_code == 200:            
        df = pd.DataFrame(columns=r.json().keys())
        df = df.append(r.json(),ignore_index=True)
        #iterating on the Json response and parsing the result
        for k, v in df["list"].iteritems():
            #print(k, v[0])        
            for j in v:
                #print("j==",j)
                res0.append(j["pop"])
                res1.append(j["main"]["temp"])
        
        idx = i[0].replace(" ","")
        #Calculating the mean 5days forecast temperature and rain
        cities_weather.append([idx,i[0],round(sum(res0)/float(len(res0)),2),round((sum(res1)/float(len(res1)))/10,2)])
                        
    else :
        print(i[0],"No found")  


outname = 'weather.csv'
outdir = './src'
if not os.path.exists(outdir):
    os.mkdir(outdir)

fullname = os.path.join(outdir, outname)  
#Storing result in a DF that will serve for the next steps
df_weather = pd.DataFrame(cities_weather , columns = ['Idx','City', 'PopAvg5d','TempAvg5d'])
df_weather.to_csv('./src/cities_weather.csv', index=False)
print("fin 2")
```

<h4>Starting the scrapping of infos about hotels from Booking.com website and store result in a temp file :src/booking.json

```python
# Import libraries

# Import os => Library used to easily manipulate operating systems
## More info => https://docs.python.org/3/library/os.html
import os 
import pandas as pd
# Import logging => Library used for logs manipulation 
## More info => https://docs.python.org/3/library/logging.html
import logging

# Import scrapy and scrapy.crawler 
import scrapy
from scrapy.crawler import CrawlerProcess
pd.options.display.max_columns = None

# Create your class WikipediaSpider(scrapy.Spider)
class bookingSpider(scrapy.Spider):
      
    liens=[]
    start_urls = [        
       #'https://www.booking.com/searchresults.fr.html?ss=Paris',
        ]
    
    for v in cities:
        start_urls.append("https://www.booking.com/searchresults.fr.html?order=bayesian_review_score&ss="+ str(v).replace(" ",""))
        
    
    # Name of your spider    
    name = "bookingSpider"
    #liens = "'https://www.booking.com/searchresults.fr.html?ss=lyon'" + " \n, " + "'https://www.booking.com/searchresults.fr.html?ss=paris'"    # Url to start your spider from 
    

    # Callback function that will be called when starting your spider
    # It will get text, author and tags of the first <div> with class="quote"
    def parse(self, response):
        
        paras = response.css('div.sr_item_content.sr_item_content_slider_wrapper') 
        #print("paras ===========================================>",paras)
        for para in paras:
            #print("para!!!!!!!! ===========================================>",para)
            yield {   
                
                'hotelName': para.css('span.sr-hotel__name::text').get(),
                'url':  para.css('div.sr-hotel__title-wrap h3 a::attr(href)').get(),
                'coordinates': para.css('div.sr_card_address_line a::attr(data-coords)').get(),
                'score': para.css('div.bui-review-score__badge::text').get(),
                'description': para.css('div.hotel_desc::text').get(),
                'city': response.url
                
        }
    
    # Name of the file where the results will be saved
filename = "booking.json"

# If file already exists, delete it before crawling (because Scrapy will 
# concatenate the last and new results otherwise)
if filename in os.listdir('src/'):
        os.remove('src/' + filename)

process = CrawlerProcess(settings = {
    'USER_AGENT': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_2) AppleWebKit/601.3.9 (KHTML, like Gecko) Version/9.0.2 Safari/601.3.9',
    'LOG_LEVEL': logging.INFO,
    "FEEDS": {
        'src/' + filename : {"format": "json"},
    }
})

# Start the crawling using the spider you defined above
process.crawl(bookingSpider)
process.start()
```

<h4>Parsing the booking.json temp file and storing data parsed( hotelname,url, description,city, gps coords) in ds_booking dataFrame

```python
ds_booking = pd.read_json('src/booking.json')
#ds_booking.head(100)
ds_booking["longitude"]=""
ds_booking["latitude"]=""

for i,t in ds_booking["coordinates"].iteritems():
    ds_booking["longitude"][i]=float(t.split(",")[0])
    ds_booking["latitude"][i]=float(t.split(",")[1])
   
    

ds_booking["hotelName"] =ds_booking["hotelName"].str.replace('\n','')
ds_booking["url"] ="https://www.booking.com"+ ds_booking["url"].str.replace('\n','')
ds_booking["description"] =ds_booking["description"].str.replace('\n','')
ds_booking["city"]= ds_booking["city"].apply(lambda i: i.split("ss=")[1])


outname = 'hotels.csv'
outdir = './src'
if not os.path.exists(outdir):
    os.mkdir(outdir)

fullname = os.path.join(outdir, outname)  

ds_booking.to_csv('./src/cities_hotels.csv', index=False)


ds_booking.head(1000)
```
